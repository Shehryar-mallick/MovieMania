<h2>importing the data set</h2>

```python
path_for_movies = 'movies.csv'
path_for_ratings = 'ratings.csv'
```

```python
# READING THE MOVIES DATA SET
movies = pd.read_csv(path_for_movies)
movies.head()
```

```python
# READING THE RATING DATASET
ratings = pd.read_csv(path_for_ratings)
ratings.head()
```

<h2>MERGING THE TWO DATASETS

```python
data = pd.merge(movies,ratings)
data.head()
```

```python
# REMOVING THE UNNECESSARY COLUMNS
data = data.drop(['genres','timestamp'],axis=1)
data
```

```python
# CREATING THE USER ITEM INTERACTION MATRIX
user_ratings_matrix = data.pivot_table(index=['userId'],columns=['title'],values='rating')
user_ratings_matrix.head()
```

```python
#DROP ALL THE MOVIES THAT HAVE NOT BEEN RATED BY EVEN 10 USERS
user_ratings_matrix = user_ratings_matrix.dropna(thresh=10,axis=1).fillna(0)
user_ratings_matrix
```

```python
#creating mean ratings data
rating = pd.DataFrame(data.groupby('title')['rating'].mean())
rating.head()
```

```python
#creating number of ratings data
rating['number_of_ratings'] = data.groupby('title')['rating'].count()
rating = rating.sort_values('number_of_ratings', ascending=False).head(10)
rating.head()
```

```python
#NOW WE CHECK THE CORELATION BETWEEN DIFFERENT MOVIES USING PEARSON CORELATION
item_similarity_df = user_ratings.corr(method='pearson')
item_similarity_df.head()
```

```python
#RECOMMENDER
def get_similar_movies(movie_name,user_rating):
    
    similar_score = item_similarity_df[movie_name]*(user_rating)
    similar_score = similar_score.to_frame(name='correlation')
    similar_score = similar_score.join(rating['number_of_ratings']).dropna()
    similar_score = similar_score[similar_score['number_of_ratings'] > 30].sort_values(by='correlation', ascending=False).iloc[1:11 , :]
    # similar_score = similar_score.sort_values(ascending=False).head(10)

    return similar_score
    
print(get_similar_movies("Forrest Gump (1994)",5))
```
